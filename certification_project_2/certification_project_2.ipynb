```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Project_Data_2.csv", index_col=0, decimal=",")
data.head()
```

```python
data.isna()
```

```python
data.dropna()
```

```python
from sklearn.decomposition import PCA
data_for_decomposition = data.iloc[:,0:]
data_for_decomposition
```

```python
model_pca = PCA(n_components=2)

pca_data = model_pca.fit(data_for_decomposition).transform(data_for_decomposition)
```

```python
print("Variance Ratio\n")
print(model_pca.explained_variance_ratio_)
print("\n\n")
```

```python
new_data = pd.DataFrame(pca_data, columns=["pca_1", "pca_2"])
new_data.index = data.index
```

```python
new_data.head()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)
```

```python
wcss = [] # wcss - within cluster squared sum of inertia

for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=10)
    kmeans.fit(new_data)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("wcss")
plt.show()
```

```python
kmeans.fit(new_data)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
new_data["cluster"] = kmeans.labels_
```

```python
new_data.head()
```

```python
import seaborn as sns
sns.lmplot('pca_1', 'pca_2', data=new_data, hue='cluster', palette='coolwarm', size=6, aspect=1, fit_reg=False)
```
